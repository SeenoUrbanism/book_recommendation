```python
import pandas as pd
import os
import random
import re
from bs4 import BeautifulSoup
import requests
```

```python
urls = [
    "https://www.nytimes.com/books/best-sellers/combined-print-and-e-book-fiction/",
    "https://www.nytimes.com/books/best-sellers/combined-print-and-e-book-nonfiction/",
    "https://www.nytimes.com/books/best-sellers/hardcover-fiction/",
    "https://www.nytimes.com/books/best-sellers/hardcover-nonfiction/",
    "https://www.nytimes.com/books/best-sellers/trade-fiction-paperback/",
    "https://www.nytimes.com/books/best-sellers/paperback-nonfiction/",
    "https://www.nytimes.com/books/best-sellers/advice-how-to-and-miscellaneous/",
    "https://www.nytimes.com/books/best-sellers/childrens-middle-grade-hardcover/",
    "https://www.nytimes.com/books/best-sellers/picture-books/",
    "https://www.nytimes.com/books/best-sellers/series-books/",
    "https://www.nytimes.com/books/best-sellers/young-adult-hardcover/",
    "https://www.nytimes.com/books/best-sellers/mass-market-monthly/",
]
    
headers = {"User-Agent": "Mozilla/5.0"}
```

```python
# we write a function to scrape ONE page
def scrape_page(url):
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, "html.parser")

    titles  = [h.get_text(strip=True) for h in soup.find_all("h3", itemprop="name")]
    authors = [
        p.get_text(strip=True).removeprefix("by ").strip()
        for p in soup.find_all("p", itemprop="author")
    ]
    return list(zip(titles, authors))
```

```python
# Loop
all_records = []
for url in urls:
    all_records.extend(scrape_page(url))
```

```python
# to one table 
df = pd.DataFrame(all_records, columns=["title", "author"])
print(df)
```

```python
df = df.drop_duplicates(subset=["title","author"])
```
