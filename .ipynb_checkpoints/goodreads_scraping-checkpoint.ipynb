```python
def extract_genre(url):
    return url.split("/shelf/show/")[-1]

def scrape_shelf_page(url):
    headers = {"User-Agent": "Mozilla/5.0"}
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, "html.parser")

    genre = url.split("/shelf/show/")[-1]

    records = []

    books = soup.select("div.elementList")
    for b in books:

        # Title
        t = b.select_one("a.bookTitle")
        title = t.get_text(strip=True) if t else None

        # Author
        a = b.select_one("span[itemprop='name']")
        author = a.get_text(strip=True) if a else None

        # Rating & year
        rating_el = b.select_one("span.greyText.smallText")
        avg_rating, rating_count, pub_year = None, None, None

        if rating_el:
            txt = rating_el.get_text(" ", strip=True)
            parts = [p.strip() for p in txt.split("—")]

            # avg rating
            if parts and "avg rating" in parts[0]:
                avg_rating = parts[0].replace("avg rating", "").strip()

            # rating count
            if len(parts) > 1 and "ratings" in parts[1]:
                rating_count = parts[1].replace("ratings", "").strip()

            # published year
            for p in parts:
                if "published" in p:
                    pub_year = p.replace("published", "").strip()

        records.append({
            "title": title,
            "author": author,
            "avg_rating": avg_rating,
            "rating_count": rating_count,
            "year_published": pub_year,
            "genre": genre               
        })

    return records
```

```python
all_data = []

for url in urls:     # urls must be a list of strings
    records = scrape_shelf_page(url)
    all_data.extend(records)

df = pd.DataFrame(all_data)
df
```

```python
dupes = df[df.duplicated(subset=["title"], keep=False)]
dupes
```

```python
genre_map = (
    df.groupby("title")["genre"]
      .apply(lambda x: list(dict.fromkeys(x)))   # keep order + remove dups
      .reset_index()
)
```

```python
genre_wide = genre_map["genre"].apply(pd.Series)
genre_wide = genre_wide.rename(columns=lambda i: f"genre{i+1}")
```

```python
base = df.drop_duplicates(subset=["title"]).drop(columns=["genre"])
```

```python
final = base.merge(
    pd.concat([genre_map["title"], genre_wide], axis=1),
    on="title",
    how="left"
)
```
