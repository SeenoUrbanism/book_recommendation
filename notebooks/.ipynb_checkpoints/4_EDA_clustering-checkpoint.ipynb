```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/clean/books_combined_genred.csv")
```

## Create TF-IDF vectors

```python
df["features_text"] = (
    df["title"] + " " +
    df["all_genres"] + " " +
    (df["author"] + " ") * 0  # <- author weighted extremely low
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    stop_words="english",
    max_features=5000  # optional, keeps matrix manageable
)

tfidf_matrix = tfidf.fit_transform(df["features_text"])
```

```python
feature_names = np.array(tfidf.get_feature_names_out())

def top_keywords_for_book(row_index, top_n=10):
    row = tfidf_matrix[row_index]
    indices = row.toarray().flatten().argsort()[-top_n:]
    return feature_names[indices][::-1]

df["top_keywords"] = [
    ", ".join(top_keywords_for_book(i, top_n=8))
    for i in range(tfidf_matrix.shape[0])
]
```

```python
df[["title", "top_keywords"]].head(10)
```

## Numerical Feature Engineering

```python
# dealying with the missing values
df["rating_count"] = df["rating_count"].fillna(0)
df["popularity"] = np.log1p(df["rating_count"])
df["year_published"] = df["year_published"].fillna(df["year_published"].median())
df["avg_rating"] = df["avg_rating"].fillna(df["avg_rating"].median())
```

```python
df["popularity"] = np.log1p(df["rating_count"])
```

### Normalize avg_rating to 0–1 range

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df["scaled_rating"] = scaler.fit_transform(df[["avg_rating"]])
```

```python
from scipy.sparse import hstack

# Fix genre column detection
genre_cols = [
    col for col in df.columns 
    if col in [
        "activism","adult","bestseller","children/young adult","fantasy","fiction",
        "guide","history","inspirational","literature","non-fiction","other",
        "poetry","romance","science","spiritual/religious","sport"
    ]
]

genre_matrix = df[genre_cols].values
numerical_matrix = df[["popularity", "scaled_rating"]].values

# Combine features
feature_matrix = hstack([tfidf_matrix, genre_matrix, numerical_matrix])

# Dimensionality reduction (PCA-like)
from sklearn.decomposition import TruncatedSVD

svd = TruncatedSVD(n_components=50, random_state=42)
reduced_features = svd.fit_transform(feature_matrix)

# KMeans clustering
from sklearn.cluster import KMeans

k = 6
kmeans = KMeans(n_clusters=k, random_state=42)
df["cluster"] = kmeans.fit_predict(reduced_features)
```

# Book Recommendation System

```python
from sklearn.metrics.pairwise import cosine_similarity

genre_cols = [
    "activism","adult","bestseller","children/young adult","fantasy","fiction",
    "guide","history","inspirational","literature","non-fiction","other",
    "poetry","romance","science","spiritual/religious","sport"
]

genre_matrix = df[genre_cols].values
genre_sim = cosine_similarity(genre_matrix)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity

# ---- Year preprocessing ----
df["year_published"] = pd.to_numeric(df["year_published"], errors="coerce")
df["year_published"] = df["year_published"].fillna(df["year_published"].median())
df["year_published"] = df["year_published"].astype(int)

# ---- Year scaling ----
scaler = MinMaxScaler()
df["year_scaled"] = scaler.fit_transform(df[["year_published"]])

# ---- Year similarity ----
year_sim = cosine_similarity(df[["year_scaled"]])

# ---- Rating preprocessing ----
df["avg_rating"] = pd.to_numeric(df["avg_rating"], errors="coerce")
df["avg_rating"] = df["avg_rating"].fillna(df["avg_rating"].median())

# ---- Rating scaling ----
scaler = MinMaxScaler()
df["rating_scaled"] = scaler.fit_transform(df[["avg_rating"]])

# ---- Rating similarity ----
rating_sim = cosine_similarity(df[["rating_scaled"]])
```

## Recommendation Function 

```python
import numpy as np

def recommend(
    query,
    df,
    cosine_sim,      # title/text similarity
    genre_sim,       # genre one-hot similarity
    rating_sim,      # rating similarity
    year_sim,        # year similarity
    n=10,
    w_title=0.45,
    w_genre=0.25,
    w_rating=0.15,
    w_year=0.10,
    w_cluster=0.05   # boost for same-cluster books
):
    """
    Unified hybrid recommendation system:
    - Fuzzy title matching
    - Title (TF-IDF) similarity
    - Genre similarity
    - Rating similarity
    - Published year similarity
    - Cluster-aware boosting
    """

    # ---------------------------------------------------------
    # 1. Fuzzy match the title
    # ---------------------------------------------------------
    q = query.lower().strip()
    matches = df[df["title"].str.lower().str.contains(q)]

    if matches.empty:
        return f"No books found for '{query}'. Try another keyword."

    # take first match
    idx = matches.index[0]
    book_cluster = df.loc[idx, "cluster"]

    # ---------------------------------------------------------
    # 2. Extract all similarity vectors for the book
    # ---------------------------------------------------------
    sim_title_vec  = cosine_sim[idx]
    sim_genre_vec  = genre_sim[idx]
    sim_rating_vec = rating_sim[idx]
    sim_year_vec   = year_sim[idx]

    # ---------------------------------------------------------
    # 3. Normalize weights (optional but improves stability)
    # ---------------------------------------------------------
    w_sum = w_title + w_genre + w_rating + w_year + w_cluster
    w_title  /= w_sum
    w_genre  /= w_sum
    w_rating /= w_sum
    w_year   /= w_sum
    w_cluster /= w_sum

    # ---------------------------------------------------------
    # 4. Compute hybrid score for all books
    # ---------------------------------------------------------
    hybrid_score = (
        w_title  * sim_title_vec +
        w_genre  * sim_genre_vec +
        w_rating * sim_rating_vec +
        w_year   * sim_year_vec
    )

    # ---------------------------------------------------------
    # 5. Apply cluster-aware boosting
    # ---------------------------------------------------------
    same_cluster = (df["cluster"] == book_cluster).astype(int)
    hybrid_score += w_cluster * same_cluster

    # ---------------------------------------------------------
    # 6. Sort by final hybrid score
    # ---------------------------------------------------------
    indices_sorted = np.argsort(hybrid_score)[::-1]

    # remove the input book itself
    indices_sorted = [i for i in indices_sorted if i != idx]

    # take top N
    top_indices = indices_sorted[:n]

    # ---------------------------------------------------------
    # 7. Return final results
    # ---------------------------------------------------------
    return df.iloc[top_indices][[
        "title",
        "author",
        "avg_rating",
        "rating_count",
        "year_published",
        "all_genres",
        "cluster"
    ]]
```

```python
recommend(
    "harry potter",
    df,
    cosine_sim,
    genre_sim,
    rating_sim,
    year_sim,
    n=10
)
```
